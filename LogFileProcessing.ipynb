# Logfile Processing

```python
import warnings
import pandas as pd
warnings.filterwarnings('ignore')
```

```python
file_name = "50EA1DAA6NA002459-1 (1).csv"
```

```python
df = pd.read_csv(file_name)

df['new_timestamp'] = df['@timestamp'].apply(lambda x: x.replace("@","").replace("+\s","").strip())
df['new_timestamp'] = pd.to_datetime(df['new_timestamp'])

# df.sort_values(by='new_timestamp', inplace=True, ignore_index=True)

print(df.shape)
df.head()
```

```python
start_at = "token received is 1"
end_at = "Waiting for the Download notification ... "
routing_activation_string = "In the routing activation"
ip_connection_string = "Successfully connected to"
request_string = "request_bytes"
response_string = "response:"
```

```python
for col in df.columns:
    n_records = len(df.loc[df[col]==start_at])
    if n_records >= 1:
        print(col, n_records)
        
for col in df.columns:
    n_records = len(df.loc[df[col]==end_at])
    if n_records >= 1:
        print(col, n_records)
        
for col in df.columns:
    n_records = len(df.loc[df[col]==ip_connection_string])
    if n_records >= 1:
        print(col, n_records)
```

```python
start_index = df[df['message']==start_at].index.tolist()
end_index = df[df['message']==end_at].index.tolist()
print(list(zip(start_index, end_index)))
```

```python
df_sub = df.loc[start_index[0]:,]
print(df_sub.shape)
df_sub.head()
```

```python
ip_df = df_sub.loc[df.message.str.contains(ip_connection_string),['message','new_timestamp']]
ip_df.reset_index(drop=True, inplace=True)
ip_df.rename(columns={'message':'connect_msg','new_timestamp':'connect_ts'},inplace=True)
print(ip_df.shape)
ip_df.head()
```
